```python
# 본문 텍스트
body_text = '봄 시즌 데일리룩으로 활용하기 좋은 화이트 오버핏 드레스 셔츠입니다. 100% 프리미엄 40수 고밀도 코튼 원단을 사용하여 착용감이 부드럽고, 구김이 적어 관리가 편안합니다. 여유 있는 드롭 숄더 패턴과 트렌디한 루즈핏 실루엣으로 제작되어 단품으로 착용하거나 스웨터, 재킷 안에 레이어드하여 연출하기 좋습니다. 베이직한 디자인에 깔끔한 봉제 마감으로 완성도를 높였으며, 슬랙스나 데님 팬츠 등 다양한 하의와 손쉽게 매치됩니다. M부터 XL까지 다양한 사이즈로 구성되어 남녀 모두 편안하게 즐기실 수 있는 고품질 베이직 아이템입니다.'

# 이미지 텍스트
image_text = 'PREMIUM COTTON 100% OVERSIZED SHIRTS / 40수 고밀도 프리미엄 원단 사용 / 드롭 숄더 루즈핏 실루엣 / 세탁 후 변형 최소화 가공 / 봄 시즌 데일리 추천 상품'

page_text = body_text+image_text
```

```python
# 질문 데이터

queries = [
    # 1. 자연스러운 질문형 (기본 데일리룩 추천)
    "봄에 청바지나 슬랙스랑 맞춰 입기 좋은 화이트 오버핏 셔츠 추천해줘.",
    
    # 2. 구체적 소재 및 핏 조건 제시형
    "100% 코튼 소재로 된 구김 적은 남성용 루즈핏 드레스 셔츠 찾아줘.",
    
    # 3. 착용 상황 및 시즌 강조형
    "봄철 데일리룩으로 레이어드해서 입기 깔끔한 흰색 오버핏 셔츠 있을까?",
    
    # 4. 키워드 중심의 자연어 검색형
    "프리미엄 40수 코튼 화이트 오버핏 셔츠 사이즈 추천 및 특징 알려줘.",
    
    # 5. 구매 목적 및 품질 비교 질문형
    "비침 적고 세탁해도 변형 적은 고품질 화이트 오버핏 셔츠 알려줘."
]
```

```python
# json_ld 데이터

json_ld_data = """
{
    "@context": "https://schema.org/",
    "@type": "Product",
    "name": "화이트 오버핏 셔츠",
    "description": "봄 시즌에 입기 좋은 100% 프리미엄 코튼 소재의 화이트 오버핏 드레스 셔츠입니다.",
    "material": "Cotton 100%",
    "color": "White"
}
"""
```

```python
# 이미지 데이터 (텍스트 추출 X)

images = [
    {"alt": "모델이 봄 화이트 오버핏 셔츠를 착용한 정면 사진입니다.", "is_text_image": True},
    {"alt": "", "is_text_image": False}
]
```

```python
# 1. 본문 텍스트 비율
# scorer.calculate_text_ratio
    # body_text_len : 본문 텍스트 길이
    # image_text_len : 이미지에서 추출한 텍스트 길이

score1 = scorer.calculate_text_ratio(
    body_text_len = len(body_text), image_text_len = len(image_text)
)
```

```python
score1
```

```python
# 2. 하이브리드 검색 모델 점수
# scorer.calculate_hybrid_search
    # user_queries : 소비자 질문 (리스트 형태)
    # page_text : 본문 전체 텍스트

score2 = scorer.calculate_hybrid_search(
    user_queries = queries,
    page_text = page_text
)
```

```python
score2
```

```python
print(score2.final_score)
print(score2.avg_lexical_overlap)
print(score2.avg_cosine_sim_raw)
```

```python
score2.query_details
```

```python
# 3. 키워드 스터핑
# scorer.calculate_keyword_stuffing
    # page_text : 본문 전체 텍스트

score3 = scorer.calculate_keyword_stuffing(page_text = page_text)
print(score3)
print('threshold 적용 전 점수:', score3.raw_score)
print('최종 점수:', score3.final_score)
```

```python
# 키워드 스터핑 예시
example_text = '봄철 하객룩 데이트룩 여친룩 여리핏 말라핏 슬림핏 하늘하늘 레이스 코튼 원피스'
score_ks_ex = scorer.calculate_keyword_stuffing(page_text = example_text)
print(score_ks_ex)
print('threshold 적용 전 점수:', score_ks_ex.raw_score)
print('최종 점수:', score_ks_ex.final_score)
```

```python
score5 = scorer.calculate_avg_alt_score(image_list=images)

# 5. 결과 확인
print(f"최종 점수 (Threshold 적용): {score5.avg_score}")
print(f"원본 평균 점수: {score5.raw_avg_score}")
print(f"통과 여부: {score5.is_valid}")
print(f"총 이미지 수: {score5.total_image_count}개")
print(f"스터핑 감지 수: {score5.stuffing_image_count}개")
```
